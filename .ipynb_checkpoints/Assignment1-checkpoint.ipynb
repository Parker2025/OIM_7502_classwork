Use NLTK to complete sentiment analysis for each of the articles

```python

tokenizer = RegexpTokenizer(r"(\w+'?\w*)")
fox_text = tokenizer.tokenize(fox_text)
reuters_text = tokenizer.tokenize(reuters_text)
```

```python
fox_text = " ".join([word for word in fox_text if word not in nltk.corpus.stopwords.words('english')])
```

```python
reuters_text  = " ".join([word for word in reuters_text if word not in nltk.corpus.stopwords.words('english')])
```

```python
sid.polarity_scores(fox_text)
```

```python
sid.polarity_scores(reuters_text)
```

Compare the results from the two articles

Overall, both of these articles scored very similarly to each other.  The negative score was slightly higher for Reuters comapared to Fox at 9.1% and 8.9% repectively.  The neutral score was 81.5% for Fox and 88.7% for Reuters, which shows that the majority of the sentiment was neutral, which for news is what most people would expect and/or want.  Lastly, I was amazed that positive sentiment in the article was for Fox at 9.6% and 2.2% for Reuters.  The one area where I'm fully shocked is the compound analysis.  Compound can be a range from -1 most negative to +1 most positive. Fox News and Reuters are on opposite ends of the sprectrum on this measure.

Using py-readability-metrics complete 2-3 different readability scores

```python
foxr = Readability(fox_text)
```

```python
reutersr = Readability(reuters_text)
```

Gunning Fog - estimates years of formal education needed to read text

```python
foxg = foxr.gunning_fog()
```

```python
print (foxg.grade_level)
```

```python
reutersg = reutersr.gunning_fog()
```

```python
print (reutersg.grade_level)
```

Coleman

```python
foxc = foxr.coleman_liau()
reutersc = reutersr.coleman_liau()
```

```python
print(foxc.grade_level)
```

```python
print(reutersc.grade_level)
```

Dale Chall - looks at firmilar words

```python
foxd = foxr.dale_chall()
reutersd = reutersr.dale_chall()
```

```python
print(foxd.grade_levels)
```

```python
print(reutersd.grade_levels)
```

I chose the metrics of Gunning Fog, Coleman Liau and Dale Chall because they are all comparing reading levels by grade level.  However, they take different approaches to get there.  Dale Chall for example looks at familiar words, while Coleman Liau and Gunning Fog both look at overall reading comprehension needed to understand while reading it once.  Although they took different methods they seemed to all point in the same direction, which for both articles was at a college level.  I'm  surprised that a news article would be at such a high reading level according to these methods.  I was anticipating it to be high school level at the highest.  I wonder if the frequency of higher education being mentioned impacts these readability scores?  According to the Coleman Liau analysis the Fox News article was at a slightly lower reading level compared to Reuters which was 21 and 23 respectively.
