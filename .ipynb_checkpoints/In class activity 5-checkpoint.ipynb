```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sb
import statsmodels.api as sm
```

```python
homes = pd.read_csv("data/Homes.csv")
```

```python
homes = pd.DataFrame(homes)
```

```python
homes.info()
```

```python
homes.head()
```

```python
homes.columns = ["Sell", "List", "Living", "Rooms", "Beds","Baths", "Age", "Acres", "Taxes"]
```

```python
homes.describe()
```

There's a lot of variance in the data specifically for sell, list and taxes.  We know this by looking at the standard deviations.  Baths, beds and acres have lower standard deviations.  This can be attributed to houses typically having a more narrow band of baths and beds, while the pricing can change drastically due to a number of characteristics.  

```python
sb.kdeplot(homes.Taxes, fill=True);
```

```python
homestrimmed = homes[homes.Taxes < 9000].copy()
```

```python
sb.kdeplot(homestrimmed.Taxes, fill=True);
```

```python
homestrimmed.describe()
```

The first thing I wanted to look at was the taxes, as this had the largest standard deviation in the dataset.  I trimmed it down to remove outliers.  I selected 9,000 as the trimming point by looking at the initial graph and seeing where the majority of the data was.  Since our data set is only 50 homes I didn't want to remove too much data.   This only removed one house from the data set.  

```python
correlation = homestrimmed.corr()
correlation['Sell'].sort_values(ascending = False)
```

Next, I wanted to look at the correlation between the selling price and other variables.  The results above show that list price and taxes are correlated.  This is to be expected, as a list price is a starting point for an offer and the taxes are based off the homes value.  The next variable with the highest correlation is Living followed by Baths, Rooms, and Beds are closely grouped together.  Age is slightly negative correlated, meaning that if you had two identical homes, but one was built 30 years ago and one was new construction, we'd predict that the newer one would sell for more. 

```python
sb.kdeplot(homestrimmed.Living, fill=True);
```

```python
correlation['Living'].sort_values(ascending = False)
```

Lastly, I wanted to explore the Living column in our dataframe some more.  As we don't have a data dictionary, I'm not certain what Living is describing. I was hoping that visualizing it, and looking at what's correlated to it, I'd be able to make an educated guess.  There's a few different ideas I have in mind. One is the living space or sq ft for the home.  The number of rooms is strongly correlated, which seems like the more rooms you have the bigger the house.  However, the units have a min of 10 and an upper trimmed limit of 39.  Perhaps this data has been normalized and that's why the numbers don't seem to be attributed to sq footage, or it could be something else entirely.   

```python
y = homestrimmed.Sell
x = homestrimmed.Taxes
```

```python
taxes_model = sm.OLS(y, x).fit()
```

```python
taxes_model.summary()
```

```python
plt.scatter(x=x, y=y, edgecolor='w')
plt.plot(x, taxes_model.predict(x), 'k');
```

Taxes are a good predictor of a houses sale price.  The R^2 value is .963, which means that 96.3% of the variance.  There's also a strong positive correlation between these two.  This means that the higher the taxes the higher the sales price would be.  Another metric to look at is p value, which is 0, which means that it's highly significant.  Overall, this a good metric to look at.  

```python
ay = homestrimmed.Age
ax = homestrimmed.Sell
homeage_model = sm.OLS(ay, ax).fit()
homeage_model.summary()
```

```python
plt.scatter(x=ax, y=ay, edgecolor='w')
plt.plot(ax, homeage_model.predict(ax), 'k');
```

This model performs significantly worse compared to the previous model.  The Rsquared value is .444, which is less than half of the taxes and sales price relationship.  I'm surprised that the trend is still slightly positive, as when I looked at the summaries it showed a negative correlation.  Perhaps this is due to the outliers that I kept in the data set.  Overall, I would not use age to predict a homes price. 

```python
homesagetrimmed = homestrimmed[homestrimmed.Age < 100].copy()
aty = homesagetrimmed.Age
atx = homesagetrimmed.Sell
homesagetrimmed_model = sm.OLS(aty, atx).fit()
homesagetrimmed_model.summary()
```
