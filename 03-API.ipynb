```python
import json
import requests
```

```python
url = "https://dummyjson.com/posts"
```

```python
request = requests.get(url).text
```

```python
data = json.loads(request)["posts"]
```

```python
f = open("data/posts.txt", "w")
json.dump(request, f)
f.close()
```
