# Linear Regression
* many implementations of OLS in various Python libraries: NumPy, SciPy, statsmodels, scikit-learn

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statsmodels.api as sm
import seaborn as sb
```

```python
housing = pd.read_csv("data/Housing.csv")
housing.head()
```

```python
sb.kdeplot(housing.price, fill=True);
```

```python
trimmed = housing[housing.price < 1e7].copy()
```

```python
sb.kdeplot(trimmed.price, fill=True);
```

```python
trimmed.describe()
```

```python
corr = trimmed.corr()
corr["price"].sort_values(ascending = False)
```

```python
y = trimmed.price / 1000
x = trimmed.area
x = sm.add_constant(x)
x
```

```python
model = sm.OLS(y, x).fit()
```

```python
model.summary()
```

```python
sample = trimmed.sample(60)
sample
```

```python
y = sample.price / 1000
x = sample.area
x = sm.add_constant(x)
```

```python
sample_model = sm.OLS(y, x).fit()
```

```python
sample_model.summary()
```

```python
sb.scatterplot(x=x["area"], y=y);
```

```python
np.sqrt(((y - model.predict(x)) **2).mean())
```

```python
model.predict(x)
```

```python
plt.scatter(x["area"], y, edgecolor="w")
plt.plot(x["area"],model.predict(x),"k");
```

```python
interecept, slope = model.params
```

### sklearn

```python
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.predict(x)
```

```python
mean_squared_error(y, reg.predict(x))**.5
```

```python
r2_score(y, reg.predict(x))
```
